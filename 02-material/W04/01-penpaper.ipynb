# task 2:

## a:
The matrix is upper triangular (all elements below the main diagonal are 0)
and since all the diagonal elements are non-zero, the determinant of the matrix is non-zero, implying that the matrix is invertible. 
Therefore, the system has a solution for any vector (b_1, b_2, b_3)

## b:
The last row is zero, making the determinant 0.
For the system to have a solution b_3 must be zero.
Therefore the system only has a solution if b_3 = 0

## c:

Therefore the system only has a solution if b_2 = b_3


# Task 3 - True or false (with a counterexample if false):

## The vectors $\mathbf{b}$ that are not in the column space $\mathbf{C}(A)$ form a subspace.
False

The set of vectors not in C(A) does not include the zero vector, since the zero vector is always in column space of A.
One of the requirements is that the subspace should always include the zero vector.

Counterexample:
A: 
(1 0)
(0 1)
The column space C(A) is all of R^2
Since there are no vectors outside C(A), there is no outside set that can be closed under addition or scalar multiplication.


## If $\mathbf{C}(A)$ contains only the zero vector, then $A$ is the zero matrix.
True

If column space(A) only has zero vector, then every column of A, must be the zero vector.


## The column space of $2A$ equals the column space of $A$.
True

Doing scalar multiplication does not change the span of its columns.
C(2A) = C(A)

## The column space of $A-I$ equals the column space of $A$ (test this).
False

As the identity matrix is not all 0, then this equals shifting the columns by the identiy matrix, thus changing the column space.
